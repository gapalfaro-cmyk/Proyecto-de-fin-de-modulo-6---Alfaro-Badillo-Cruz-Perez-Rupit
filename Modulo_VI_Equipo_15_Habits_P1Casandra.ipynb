```python
# tolist para que nos de una lista porque en el anterior código nos daba index
numeric_features = df.select_dtypes(include=np.number).columns.tolist()

categorical_features = df.select_dtypes(include="object").columns.tolist()

print("Variables numéricas:")
print(numeric_features)

print()

print("Variables categóricas:")
print(categorical_features)
```

```python
# empieza preprocesamiento
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

imputamos los valores faltantes de parental_education_level con:
imputer = SimpleImputer(strategy="most_frequent")
pero primero vemos cuáles son las categorías existentes y las más frecuentes

```python
df["parental_education_level"].value_counts(dropna=False)
```

La variable parental_education_level presentó valores faltantes. Debido a que se trata de una variable categórica y el porcentaje de datos ausentes fue bajo (≈9 %), los valores faltantes se imputaron utilizando la categoría más frecuente (most_frequent). Esta estrategia permitió conservar todas las observaciones del conjunto de datos sin introducir una nueva categoría artificial.

```python
numeric_transformer = Pipeline(steps=[
    ("scaler", StandardScaler())
])

numeric_transformer
```

```python
#imputaremos parental_education_level
categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)
preprocessor
```

```python
#Separar las variables numéricas
#Escala las numéricas
#Separa las categóricas
#imputa los Nan de parental_education_level
#One-Hot Encoding
#Une nuevamente todas las columnas
X_preprocessed = preprocessor.fit_transform(df)
```

```python
print(type(X_preprocessed))
print(X_preprocessed.shape)
```

```python
feature_names = preprocessor.get_feature_names_out()

print(feature_names)
```

### Aprendizaje no supervisado

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
#fit: aprende cómo combinar las 25 variables
#transform: transforma cada estudiante a las nuevas componentes principales
X_pca = pca.fit_transform(X_preprocessed)
```

```python
print(type(X_pca))
print(X_pca.shape)
```

```python
#Ver cuánta varianza explica cada componente
explained_variance = pca.explained_variance_ratio_

explained_variance
```

```python
#calcula la varianza acumulada
cumulative_variance = np.cumsum(explained_variance)

cumulative_variance
```

```python
#scree plot
plt.figure(figsize=(8,5))

plt.plot(
    range(1, len(cumulative_variance)+1),
    cumulative_variance,
    marker="o"
)

plt.xlabel("Número de componentes principales")
plt.ylabel("Varianza acumulada")
plt.title("Varianza explicada acumulada")

plt.grid(True)

plt.show()
```

```python
#13 componentes conservan aproximadamente el 91 % de la información
pca = PCA(n_components=13)

X_pca = pca.fit_transform(X_preprocessed)

print(X_pca.shape)
```

```python
#pca con 2 componentes
#la componente principal1 pca1 explica 15.95% de la variabilidad
# Componente principal 2 (PC2) explica 8.90% de la variabilidad
pca_2 = PCA(n_components=2)

X_pca_2 = pca_2.fit_transform(X_preprocessed)

pca_2.explained_variance_ratio_
```

```python
#cuánto explica la representación en 2 dimensiones
#24.85% de la varianza total
pca_2.explained_variance_ratio_.sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    X_pca_2[:,0],
    X_pca_2[:,1]
)

plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")

plt.title("Proyección de estudiantes mediante PCA")

plt.grid()

plt.show()
```

No se observan grupos claramente separados.
La mayoría de los estudiantes se concentran cerca del centro.
Hay algunos estudiantes alejados (posibles observaciones atípicas), pero no forman un grupo propio.
Esto es consistente con que los dos primeros componentes sólo explican el 24.85% de la varianza. Es decir, al reducir de 13 dimensiones a 2 se pierde bastante información y por eso la separación entre observaciones no es muy evidente.

La proyección mediante PCA muestra una distribución continua de los estudiantes sin agrupaciones claramente definidas. Los dos primeros componentes principales explican el 24.85% de la varianza total, por lo que la representación bidimensional conserva únicamente una parte de la información del conjunto de datos.

## t-SNE

```python
from sklearn.manifold import TSNE
# crear modelo
tsne = TSNE(
    n_components=2,
    random_state=42
)
```

```python
#transformando
# el resultado nos dice que cada estudiante es representado en 2 coordenadas
X_tsne = tsne.fit_transform(X_preprocessed)
X_tsne.shape
```

```python
#grafica t-SNE
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    X_tsne[:,0],
    X_tsne[:,1]
)

plt.xlabel("Dimensión t-SNE 1")
plt.ylabel("Dimensión t-SNE 2")

plt.title("Proyección de estudiantes usando t-SNE")

plt.grid()

plt.show()
```

regiones con mayor concentración de puntos;
zonas menos densas;
una estructura mucho más compleja que la observada con PCA

La proyección obtenida mediante t-SNE revela una estructura más compleja que PCA. Se observan regiones de mayor densidad y cierta organización local entre las observaciones, aunque no aparecen grupos completamente separados. Esto sugiere que existen similitudes locales entre estudiantes que PCA no logra mostrar con la misma claridad

En t-SNE no interpretamos la varianza explicada como en PCA. t-SNE no está tratando de conservar la varianza, sino la estructura de vecindades

##Otras conclusiones

Que no hayan salido grupos claros en PCA, t‑SNE o UMAP depende de la naturaleza de los datos:

🔹 t‑SNE y UMAP: sí suelen mostrar agrupamientos locales, pero si los hábitos estudiantiles son muy variados y se mezclan, es normal que no aparezcan “islas” bien definidas.

🔹 Interpretación correcta: los resultados muestran una distribución continua de hábitos, con algunas concentraciones locales pero sin clusters totalmente separados. Eso en sí es un hallazgo: significa que los estudiantes no se dividen en grupos rígidos, sino que sus hábitos forman un espectro.

los métodos muestran que los hábitos no forman clusters definidos, sino una distribución continua con similitudes locales. Revela que el comportamiento estudiantil no se puede encasillar en categorías fijas.

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
labels_kmeans = kmeans.fit_predict(X_preprocessed)

# Clustering aglomerativo
agglo = AgglomerativeClustering(n_clusters=3)
labels_agglo = agglo.fit_predict(X_preprocessed)

# Visualización con PCA (solo para graficar en 2D)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_preprocessed)

plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_kmeans)
plt.title("Clusters por K-Means")
plt.show()

plt.scatter(X_pca[:,0], X_pca[:,1], c=labels_agglo)
plt.title("Clusters por método aglomerativo")
plt.show()
```
